```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import glob
import matplotlib.image as mimg
```

```python
def fr(image):
    f = np.fft.fft2(image)
    fshift = np.fft.fftshift(f)
    specturm = np.log(np.abs(fshift))
    return specturm
```

```python
rasm1 = cv2.imread('C:/Users/Lucas/Desktop/harflar/A1.jpg')
rasm2 = cv2.imread('C:/Users/Lucas/Desktop/harflar/A2.jpg')
rasm1 = cv2.cvtColor(rasm1,cv2.COLOR_BGR2GRAY)
rasm2 = cv2.cvtColor(rasm2,cv2.COLOR_BGR2GRAY)
```

```python
rasm1f = fr(rasm1)
rasm2f = fr(rasm2)

plt.figure(figsize=(20,10))
plt.subplot(121)
plt.imshow(rasm1f,cmap = 'gray')
plt.subplot(122)
plt.imshow(rasm2f,cmap = 'gray')
plt.show()
```

```python
rasms = cv2.imread('rasm/number.jpg')
rasms = cv2.cvtColor(rasms,cv2.COLOR_BGR2RGB)
edge = cv2.Canny(rasms,100,120)
```

```python
line_rasms = np.copy(rasms)
line = cv2.HoughLinesP(edge,rho=1,theta=np.pi/180,threshold=30,minLineLength=25,maxLineGap=3)
for i in line:
    for a,b,a1,b1 in i:
        cv2.line(line_rasms,(a,b),(a1,b1),(0,0,255),25)
plt.imshow(line_rasms)
plt.show()
```
